## Data Visualization

Data visualization is the technique to present the data in a pictorial or graphical format.

It has various advantages:
* It analyzes and explores large data easily.
* It simplifies complex quantitative information.
* It identifies the areas for improvement.
* It identifies relationships between data points and variables.
* It explores new patterns and reveals hidden patterns.

### 1. Scatter Plot

A scatterplot displays the relationship between 2 numeric variables. For each data point, the value of its first variable is represented on the X axis, the second on the Y axis.

|Merits|Demerits|
|:--------------:|:------------:|
|It is easy to plot.|It is not suitable for large datasets.|
|It is easy to understand and interpret.|For large dataset, analysing a trend becomes difficult.|
|It helps in finding outliers.|It takes numerical data only.|
|It helps in finding correlation between two variables - strong, weak, positive, negative, etc.|It is used in bivariate analysis only.|

```python
sns.scatterplot(df["sepal_length"],df["petal_length"])
plt.figure(figsize=(8,6))
```

```python
a = sns.load_dataset("titanic")
sns.scatterplot(a["age"],a["fare"])
```

### 2. Bar Plot

A barplot is one of the most common types of graphic. It shows the relationship between a numerical and a categorical variable. Each entity of the categorical variable is represented as a bar. The size of the bar represents its numeric value.

|Merits|Demerits|
|:--------------:|:------------:|
|It is easy to plot.|It sometimes becomes difficult to interpret and thus requires additional written or verbal explanation.|
|It helps in comparing two or more categories.|It is not suitable for large number of categories.|
|It is easy to understand.|It takes categorical data.|
|It is used in bivariate and multivariate analysis.|It doesn't tell about outliers in the data.|

```python
sns.barplot(df["petal_width"],df["species"])
plt.figure(figsize=(5,5))
```

### 3. Histogram

A histogram takes as input a numerical variable only. The variable is cut into several bins (or intervals) and the number of observation per bin is represented by the height of the bar.

|Merits|Demerits|
|:--------------:|:------------:|
|It gives distribution of data by displaying the number of values present in a specified interval.|It needs continuous data.|
|It is suitable for large dataset as data can be grouped into intervals.|As the data is grouped into intervals, it is difficult to calculate precise central tendency.|
|It is easy to understand and interpret.|Comparing two datasets using histograms is very challenging.|
|It tells about skewness of the data.|It is used in univariate analysis.|

Don’t confuse it with a barplot. A barplot gives a value for each group of a categorical variable and here, we have only a numerical variable and we check its distribution.

```python
sns.histplot(df["sepal_length"])
plt.figure(figsize=(5,5))
```

```python
sns.histplot(a["fare"])
```

### 4. Lineplot

A line plot displays the trend or pattern of one or several numeric variables. Data points are connected by straight line segments. It is similar to a scatter plot except that the data points are ordered (typically by their x-axis value) and joined with straight line segments. A line plot is often used to visualize a trend in data over intervals of time – a time series – thus the line is often drawn chronologically.

|Merits|Demerits|
|:--------------:|:------------:|
|It is simple, easy to understand and interpret.|It is ideal for representing numerical data only.|
|It is helpful in analysing trends overtime. It can measure and show small changes that are difficult to show by other graphs.|It is ideal to represent continuous data.|
|It allows to make estimates of missing data and possible extrapolation of data.|To plot multiple lineplots on a single graph, we need consistent scales.|
|It is used in bivariate and multivariate analysis.|It does not tell about skewness of the data.|

Dual axis with different scaes may lead to incorrect data interpretations.

```python
sns.lineplot(df["petal_length"],df["petal_width"])
```

```python
sns.lineplot(x = df["petal_length"],)
```

### 5. Heatmap

A heatmap is a graphical representation of data where the individual values are represented as colors. It is useful to display a general view of numerical data and not to extract specific data point. It is also useful to display the result of hierarchical clustering.

|Merits|Demerits|
|:--------------:|:------------:|
|It uses entire dataset to draw meaningful insights.|It represents numerical data only.|
|It gives correlation between different columns of the data.|It is difficult to interpret data using heatmap. Different persons can derive different interpretations.|
|It can deal with large datasets.|It does not tell about outliers or missing values in the data.|
|It is used in multivariate analysis.|Heatmap could be misleading as it often involves large amount of data.|

```python
sns.heatmap(df.iloc[:,:-1])
```

### 6. Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary - “minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”. 

|Merits|Demerits|
|:--------------:|:------------:|
|Box plot summarizes the variation in data and display the result in a single graph. |It represents numerical data only.|
|It can tell whether the data is symmetrical or not, or if it is skewed.|Box plot is not relevant for detailed analysis of the data as it deals with a summary of the data distribution.|
|Boxplots can tell us about outliers and what their values are.|It is not useful for large datasets.|
|It is effective and easy to read.|It is a useful tool for univariate analysis only.|

```python
sns.boxplot(df["petal_length"])
```

```python
sns.boxplot(a["age"])
```

### 7. Pie Chart
A pie chart is a circle divided into sectors, each representing a proportion of the whole. It is often used to show percentage, where the sum of the sectors equals 100%.

|Merits|Demerits|
|:--------------:|:------------:|
|It summarizes large data in a visual form. |It requires explanation and theoretical knowledge about circle, central angle, etc.|
|It displays mupltiple classes of data in one chart.|When there are too many or similar sized slices, pie chart becomes less effective.|
|It is more visually appealing than other graphs.|It doesn't reveal exact values.|
|It is helpful in understanding data composition.|It is difficult to compare two datasets.|

```python
a = sns.load_dataset("titanic")

a1 = a["embark_town"].value_counts().rename_axis("Departure").reset_index(name = "Counts")
a1.head()
```

```python
labels = a1["Departure"]
values = a1["Counts"]

plt.pie(values, labels=labels, autopct='%1.2f%%')
plt.figure(figsize=(5,5))
```

### 8. Pairplot

Pairplot is a plot of multiple pairwise bivariate distributions in a dataset.

|Merits|Demerits|
|:--------------:|:------------:|
|It gives relationship between all numerical columns in the data in a single graph. |It does not plot categorical variables.|
|It gives univariate analysis along main diagonal through histograms.|When there are too many numerical columns, pairplot becomes crowded and diffcult to analyse.|
|It saves memory by providing multiple plots in a single graph through a single code.|It uses only scatterplots and histograms.|
|It can do both univariate and bivariate analysis.|It is not useful for multivariate analysis.|

```python
sns.pairplot(df)
```

### 9. Distplot

A distplot or distribution plot gives a univariate distribution of observations.

|Merits|Demerits|
|:--------------:|:------------:|
|It gives data distribution of a variable against the density distribution.|It does not plot categorical variables.|
|It displays the distribution and range of data values over a scale.|It does not give a detailed account of the ranging and distribution of values as it dealswith summary of the data distribution.|
|It tells about skewness of the data.|It does only univariate analysis.|

```python
sns.distplot(df["sepal_length"])
```

### 10. KDE Plot

KDE plot or Kernel Density Estimation plot depicts the probability density function of the continuous and non-parametric data variables.

|Merits|Demerits|
|:--------------:|:------------:|
|It produces smooth estimate of the probability distribution function.|It does not plot categorical variables.|
|It gives a plot that is less clustered and more interpretable.|If there are too many peaks, it makes the distribuion difficult to interpret.|
|It uses all the data points and more convincingly suggest multimodality.|The data representation is poor via KDE plot. The data is represented vaguely and causes disruptions.|
|It can do both univariate and bivariate analysis.|It is always biased at the boundaries when the data is bounded.|

```python
sns.kdeplot(df["petal_length"])
plt.figure(figsize=(8,6))
```

### 11. Bubble Chart

A bubble chart is a variation of a scatter chart in which the data points are replaced with bubbles, and an additional dimension of the data is represented in the size of the bubbles. Just like a scatter chart, a bubble chart does not use categorical values — both horizontal and vertical axes show numerical values.

|Merits|Demerits|
|:--------------:|:------------:|
|Bubble charts illustrate three different variables, making them uniquely valuable in displaying certain types of information.|It is difficult to read and understand.|
|Bubble charts can make it easy to digest a relatively large amount of information.|Due to circle sizes, it becomes difficult to ascertain actual values.|
|They are a way to tell a story visually and can provide a quick assessment of the relationship between different datasets.|It cannot be used for large datasets.|
|Well-designed bubble charts effectively display three fields of data, using bubble position and proportion.|It uses only numeric data.|
|It is dynamic enough to analyze finance, sales and marketing scenarios beyond time-series.|The overlapping of bubbles can make it difficult to distinguish them.|
|It helps to correlate data, visualize patterns and uncover trends.|It is difficult to depict zero or negative values.|

```python
sns.scatterplot(df["sepal_width"], df["petal_length"], size = df["petal_width"], sizes = (20,400), hue = df["species"])
plt.legend(bbox_to_anchor=(1.5,1))
```

### 12.  Violin Plot

Violin plot allows to visualize the distribution of a numeric variable for one or several groups. Each violin represents a group or a variable. The shape represents the density estimate of the variable - the more the data points in a specific range, the larger the violin is for that range. 

|Merits|Demerits|
|:--------------:|:------------:|
|Violin plot is a powerful data visualization technique since it allows to compare both the ranking of several groups and their distribution.|It is used to visualize the distribution of numerical data only.|
|It is similar to a boxplot, but allows a deeper understanding of the distribution.|It is used in univariate analysis only.|
|It is suitable for large datasets.|Its meaning can be harder to grasp for many readers.| 
|Unlike a box plot that can only show summary statistics, violin plots depict summary statistics as well as the density of each variable.|It doesn't show outliers.|

```python
plt.figure(figsize=(4,4))
sns.violinplot(df["sepal_length"])
```

### 13. Area Chart

An area chart is similar to a line chart and represents the evolution of a numeric variable. Basically, the X-axis represents an ordered variable and the Y axis gives the value of another variable. Data points are connected by straight line segments and the area between the X-axis and the line is colored or shaded.

|Merits|Demerits|
|:--------------:|:------------:|
|An area chart graphically displays quantitative data.|Removing the area under the curve would make it a line chart which conveys the same information.|
|It can show how one or more groups’ numeric values change over the progression of a second variable.|It needs numeric variables.|

```python
x = range(1,151)
plt.stackplot(x, df["sepal_length"],df["petal_length"], labels = ["sepal_length","petal_length"])
plt.legend(loc='upper left')
```

### 14. Facet Grid

A FacetGrid is a multi-axes grid with subplots visualizing the distribution of variables of a dataset and the relationship between multiple variables. 

```python
b = sns.load_dataset("tips")

plot = sns.FacetGrid(b, col = "day")
plot.map(sns.scatterplot, "tip", "total_bill")
plt.show()
```

### 15. Catplot

A catplot shows frequencies of the groups of categorical variables. 
It is a type of scatterplot where categorical data is represented.

```python
b = sns.load_dataset("tips")

sns.catplot(data = b, x ="day", y ="tip")
plt.figure(figsize=(3,3))
plt.show()
```

### 16. Count plot

A count plot shows the counts of observations in each categorical bin using bars.

```python
a = sns.load_dataset("titanic")

sns.countplot(a["embark_town"], hue = a["alive"])
plt.show()
```
